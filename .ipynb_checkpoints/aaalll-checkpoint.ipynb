```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def is_passed(self):
        return self.marks > 40


s1 = Student("Alice", 55)
s2 = Student("Bob", 35)

if s1.is_passed():
    print(f"{s1.name} has passed.")
else:
    print(f"{s1.name} has failed.")

if s2.is_passed():
    print(f"{s2.name} has passed.")
else:
    print(f"{s2.name} has failed.")
```

```python
class Employee:
    company_name = "TechCorp"

    def __init__(self, name):
        self.name = name

    @classmethod
    def change_company(cls, new_name):
        cls.company_name = new_name


emp1 = Employee("Alice")
emp2 = Employee("Bob")

print(f"{emp1.name} works at {emp1.company_name}")
print(f"{emp2.name} works at {emp2.company_name}")

Employee.change_company("CV CORP ")

print(f"{emp1.name} works at {emp1.company_name}")
print(f"{emp2.name} works at {emp2.company_name}")

```

```python
class MathOps:
    @staticmethod
    def is_even(num):
        return num % 2 == 0

print(MathOps.is_even(10))
m = MathOps()
print(m.is_even(7))
```

```python
class Car:
    wheels = 4
    def __init__(self, mileage):
        self.mileage = mileage
    def display_specs(self):
        print(self.mileage)
        print(self.wheels)
    @classmethod
    def change_wheels(cls, wheels):
        cls.wheels = wheels

obj = Car(100)
obj.display_specs()
Car.change_wheels(3)
obj.display_specs()
```

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    @staticmethod
    def to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

    def show_conversion(self):
        print("Celsius:", self.celsius)
        print("Fahrenheit:", Temperature.to_fahrenheit(self.celsius))

t = Temperature(25)
t.show_conversion()
```

```python
class Book:
    total_books = 0
    def __init__(self, title, author):
        self.title = title
        self.author = author
        Book.total_books += 1

    @classmethod
    def from_string(cls, book_str):
        title, author = book_str.split('-')
        if cls.is_valid_title(title):
            return cls(title, author)
        else:
            print("Invalid title. Book not created.")

    @staticmethod
    def is_valid_title(title):
        return len(title.strip()) >= 3


b1 = Book("Python", "Guido")
b2 = Book.from_string("AI-John")
b3 = Book.from_string("A-Jane")
```

```python
class Employee:
    bonus_rate = 0.1

    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def final_salary(self):
        return self.base_salary + (Employee.bonus_rate * self.base_salary)

    @classmethod
    def updatebonus(cls, new_rate):
        cls.bonus_rate = new_rate

    @staticmethod
    def is_valid_salary(salary):
        return salary > 0


emp1 = Employee("John", 20000)
emp2 = Employee("Alice", 30000)

print(emp1.name, emp1.final_salary())
print(emp2.name, emp2.final_salary())

emp1.updatebonus(0.2)

print("After updating bonus:", emp1.bonus_rate)
print(emp1.name, emp1.final_salary())
print(emp2.name, emp2.final_salary())


```

```python
class Course:
    total_students = 0

    def __init__(self, student_name):
        self.student_name = student_name

    def enroll(self):
        Course.total_students += 1
        print(f"{self.student_name} has been enrolled.")

    @classmethod
    def show_total(cls):
        print(f"Total students enrolled: {cls.total_students}")

    @staticmethod
    def is_eligible(age):
        return age >= 18


s1 = Course("Ravi")
s2 = Course("Sita")
s3 = Course("Arjun")

if Course.is_eligible(19):
    s1.enroll()
if Course.is_eligible(17):
    s2.enroll()
if Course.is_eligible(20):
    s3.enroll()

Course.show_total()
```

```python
class BankAccount:
    bank_name = "State Bank"

    def __init__(self, holder, balance=0):
        self.holder = holder
        self.balance = balance

    def deposit(self, amount):
        if BankAccount.validate_amount(amount):
            self.balance += amount
            print(f"{self.holder} deposited {amount}. New balance: {self.balance}")
        else:
            print("Invalid deposit amount.")

    @classmethod
    def change_bank_name(cls, new_name):
        cls.bank_name = new_name
        print(f"Bank name changed to {cls.bank_name}")

    @staticmethod
    def validate_amount(amount):
        return amount > 0


a1 = BankAccount("Ravi", 1000)
a2 = BankAccount("Sita", 500)

print(f"Initial Bank: {BankAccount.bank_name}")
a1.deposit(500)
a2.deposit(-200)

BankAccount.change_bank_name("National Bank")

print(f"\nAfter Bank Name Change:")
print(f"{a1.holder}'s Bank: {BankAccount.bank_name}, Balance: {a1.balance}")
print(f"{a2.holder}'s Bank: {BankAccount.bank_name}, Balance: {a2.balance}")
```

```python
class Student:
    passing_marks = 40

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def result(self):
        if self.marks >= Student.passing_marks:
            print(f"{self.name} has passed.")
        else:
            print(f"{self.name} has failed.")

    @classmethod
    def update_passing_marks(cls, new_marks):
        cls.passing_marks = new_marks
        print(f"Passing marks updated to {cls.passing_marks}")

    @staticmethod
    def grade_category(marks):
        if marks >= 85:
            return "A"
        elif marks >= 70:
            return "B"
        else:
            return "C"


s1 = Student("Ravi", 78)
s2 = Student("Sita", 35)

print("Initial Results:")
s1.result()
s2.result()

Student.update_passing_marks(50)

print("\nAfter Updating Passing Marks:")
for student in [s1, s2]:
    grade = Student.grade_category(student.marks)
    print(f"{student.name} scored {student.marks} → Grade: {grade}")
    student.result()
```
